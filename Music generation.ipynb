```python
notes = []

for file in glob.glob("midi_songs/*.mid"):
    midi = converter.parse(file)
    elements_to_parse = midi.flat.notes
    
    for ele in elements_to_parse:
        if isinstance(ele,note.Note):
            notes.append(str(ele.pitch))
        elif isinstance(ele,chord.Chord):
            notes.append("+".join(str(n) for n in ele.normalOrder))
```

```python
with open("notes","wb") as f:
    pickle.dump(notes,f)
```

```python
with open("notes","rb") as f:
    notes = pickle.load(f)
```

```python
n_vocab = len(set(notes))
n_vocab
```

## Prepare sequential data for LSTM

```python
sequence_len = 100
```

```python
pitchnames = sorted(set(notes))
```

```python
ele_to_int = dict((ele,num) for num,ele in enumerate(pitchnames))
ele_to_int
```

```python
network_input = []
network_output = []

for i in range (len(notes)-sequence_len):
    seq_in = notes[i:i+sequence_len]
    seq_out = notes[i+sequence_len]
    
    network_input.append([ele_to_int[ele] for ele in seq_in])
    network_output.append(ele_to_int[seq_out])
```

```python
len(network_input[0])
```

```python
n_patterns = len(network_input)
n_patterns
```

```python
network_input = np.reshape(network_input,(n_patterns,sequence_len,1))
network_input.shape
```

```python
normalised_network_input = network_input/float(n_vocab)
```

```python
network_output = np_utils.to_categorical(network_output)
network_output.shape
```

## Model Architecture

```python
from keras.models import Sequential,load_model
from keras.layers import *
from keras.callbacks import ModelCheckpoint,EarlyStopping
```

```python
model = Sequential()
model.add(LSTM(units = 512,input_shape = (100,1),return_sequences = True))
model.add(Dropout(0.3))
model.add(LSTM(512,return_sequences = True))
model.add(Dropout(0.3))
model.add(LSTM(512))
model.add(Dense(256))
model.add(Dense(n_vocab,activation="softmax"))
```

```python
model.compile(loss="categorical_crossentropy",optimizer="adam")
model.summary()
```

```python
checkpoint = ModelCheckpoint("model.hdf5",monitor="loss",verbose=0,save_best_only=True,mode="min")
model_his = model.fit(normalised_network_input,network_output,epochs=5,batch_size=64,callbacks=[checkpoint])
```

```python
model = load_model("new_weights.hdf5")
```

## Predictions 

```python
network_input = []
for i in range(len(notes)-sequence_len):
    seq_in = notes[i:i+sequence_len]
    network_input.append([ele_to_int[c] for c in seq_in])
```

```python
start = np.random.randint(len(network_input)-1)

int_to_ele = dict((num,ele) for num,ele in enumerate(pitchnames))

pattern = network_input[start]
prediction_output = []

for i in range(100):
    prediction_input = np.reshape(pattern,(1,len(pattern),1))
    prediction_input = prediction_input/float(n_vocab)
    
    prediction = model.predict(prediction_input,verbose=0)
    idx = np.argmax(prediction)
    prediction_output.append(int_to_ele[idx])
    pattern.append(idx)
    pattern = pattern[1:]
```

```python
print(prediction_output)
```

## Creating midi file

```python
offset = 0
output_notes = []
for pattern in prediction_output:
    if('+' in pattern or pattern.isdigit()):
        notes_in_chord = pattern.split('+')
        temp_notes = []
        for current_note in notes_in_chord:
            new_note = note.Note(int(current_note))
            new_note.storedInstrument = instrument.Piano()
            temp_notes.append(new_note)
            
        new_chord = chord.Chord(temp_notes)
        new_chord.offset = offset
        output_notes.append(new_chord)
    
    else:
        new_note = note.Note(pattern)
        new_note.offset = offset
        new_note.storedInstrument = instrument.Piano()
        output_notes.append(new_note)
    
    offset += 0.5
```

```python
midi_stream = stream.Stream(output_notes)
midi_stream.write('midi',fp="test_output.mid")
```
